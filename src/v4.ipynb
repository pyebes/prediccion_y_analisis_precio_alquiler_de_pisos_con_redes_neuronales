```python
import tensorflow as tf
import pandas as pd

import numpy as np

print(tf.__version__)

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
df_v1 = pd.read_excel('../data/testing_v1.xlsx', index_col='ID')
df_v2 = pd.read_excel('../data/testing_v2.xlsx', index_col='ID')
df_v3 = pd.read_excel('../data/testing_v3.xlsx', index_col='ID')

testing_completo = pd.read_excel('../data/testing_completo.xlsx')
```

```python
from tensorflow.keras.models import load_model

model_v1 = load_model('models/model_v1.h5')
model_v2 = load_model('models/model_v2.h5')
model_v3 = load_model('models/model_v3.h5')
```

```python
x_v1 = df_v1.iloc[:,:-1]
y_v1 = df_v1.iloc[:,-1]

x_v2 = df_v2.iloc[:,:-1]
y_v2 = df_v2.iloc[:,-1]

x_v3 = df_v3.iloc[:,:-1]
y_v3 = df_v3.iloc[:,-1]
```

```python
print("Shape of x_v1:", x_v1.shape)
print("Expected input shape for model_v1:", model_v1.input_shape)

print("Shape of x_v2:", x_v2.shape)
print("Expected input shape for model_v2:", model_v2.input_shape)

print("Shape of x_v3:", x_v3.shape)
print("Expected input shape for model_v3:", model_v3.input_shape)
```

```python
predict_v1 = model_v1.predict(x_v1)
predict_v2 = model_v2.predict(x_v2)
predict_v3 = model_v3.predict(x_v3)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

print('Métricas predicción modelo 1:')
rmse_test = np.sqrt(mean_squared_error(predict_v1, y_v1))
print("RMSE (prueba):", rmse_test)
r2_test = r2_score(predict_v1, y_v1)
print("R^2 (prueba):", r2_test)

print('Métricas predicción modelo 2:')
rmse_test = np.sqrt(mean_squared_error(predict_v2, y_v2))
print("RMSE (prueba):", rmse_test)
r2_test = r2_score(predict_v2, y_v2)
print("R^2 (prueba):", r2_test)

print('Métricas predicción modelo 3:')
rmse_test = np.sqrt(mean_squared_error(predict_v3, y_v3))
print("RMSE (prueba):", rmse_test)
r2_test = r2_score(predict_v3, y_v3)
print("R^2 (prueba):", r2_test)
```

```python
df_results = pd.DataFrame({'ID': testing_completo['ID'], 'Precio alquiler al año': testing_completo['Precio alquiler al año'],'Prediccion precio alquiler aplicado al año': predict_v1.flatten(), 'Prediccion gastos generales al año': predict_v2.flatten(), 'Prediccion gastos iniciales': predict_v3.flatten(), 'Años alquiler': testing_completo['Años contrato alquiler']})

df_results.set_index(df_results['ID'], inplace=True)
df_results.drop(columns='ID',inplace=True)
```

```python
df_results[['Prediccion precio alquiler aplicado al año', 'Prediccion gastos generales al año',	'Prediccion gastos iniciales']] = df_results[['Prediccion precio alquiler aplicado al año', 'Prediccion gastos generales al año',	'Prediccion gastos iniciales']].astype('float64').round(2)
```

```python
df_results.head(5)
```

# Calculamos diferentes valores para ordenar los pisos del más óptimo al que menos

---
# 1) Beneficio neto anual

    Beneficio neto anual = precio alquiler aplicado al año (predichos) - precio alquiler al año - gastos generales al año (predichos)

```python
df_results['Beneficio neto anual'] = (df_results['Prediccion precio alquiler aplicado al año'] - df_results['Precio alquiler al año'] - df_results['Prediccion gastos generales al año']).round(2)
```

```python
df_results[['Beneficio neto anual']].sort_values(by='Beneficio neto anual', ascending=False).head(5)
```

---
# 2) Beneficio total: (teniendo en cuenta la inversion inicial)

    Beneficio total = (precio alquiler aplicado al año (predicho) - precio alquiler al año - gastos generales al año (predicho)) * años contrato - inversion inicial (predicho)

    Beneficio total = Beneficio al año (sin contar inversion inicial) * años contrato - inversion inicial (predicho)

```python
df_results['Beneficio total'] = df_results['Beneficio neto anual']*df_results['Años alquiler'] - df_results['Prediccion gastos iniciales']
```

```python
df_results[['Beneficio total']].sort_values(by='Beneficio total', ascending=False).head(5)
```

---
# 3) Beneficio al año (contando inversion inicial):

    Beneficio al año (contando inversion inicial) = Beneficio total / Años contrato

```python
df_results['Beneficio al año (contando inversion inicial)'] = (df_results['Beneficio total'] / df_results['Años alquiler']).round(2)
```

```python
df_results[['Beneficio al año (contando inversion inicial)']].sort_values(by='Beneficio al año (contando inversion inicial)', ascending=False).head(5)
```

---
# 4) Periodo de recuperacion de la inversion (payback period)

    periodo de recuperacion = gastos iniciales (predichos) / beneficio neto anual

```python
df_results['Periodo de recuperacion'] = (df_results['Prediccion gastos iniciales'] / df_results['Beneficio neto anual']).round(3)
```

```python
df_results[['Periodo de recuperacion']][df_results['Periodo de recuperacion']>0].sort_values(by='Periodo de recuperacion', ascending=True).head(5)
```

---
# 5) Retorno sobre la inversion (ROI)

    ROI anual = (Beneficio neto anual / gastos iniciales (predichos)) * 100

```python
df_results['ROI'] = ((df_results['Beneficio neto anual']/df_results['Prediccion gastos iniciales'])*100).round(3)
```

```python
df_results[['ROI']].sort_values(by='ROI', ascending=False).head(5)
```

---
# 6) VAN

    VAN = (años contrato * Beneficio neto anual)/(1+r)^t - gastos iniciales (predichos)

```python
import numpy_financial as npf

tasa_descuento = 0.075

def calcular_flujos_de_caja(inversion_inicial, beneficio_neto_anual, anos_contrato):
    return [-inversion_inicial] + [beneficio_neto_anual] * int(anos_contrato)

df_results['Flujos de caja'] = df_results.apply(lambda row: calcular_flujos_de_caja(row['Prediccion gastos iniciales'], row['Beneficio neto anual'], row['Años alquiler']), axis=1)

df_results['VAN'] = (df_results['Flujos de caja'].apply(lambda flujos: npf.npv(tasa_descuento, flujos))).round(2)
```

```python
df_results[['VAN']].sort_values(by='VAN', ascending=False).head(5)
```

---
# 7) Tasa Interna de Retorno (TIR), la TIR que hace que el VAN sea 0.

    VAN = 0 = (años contrato * Beneficio neto anual)/(1+IRR)^t - gastos iniciales (predichos)

```python
df_results['TIR'] = (df_results['Flujos de caja'].apply(lambda flujos: npf.irr(flujos))).round(3)
```

```python
df_results[['TIR']].sort_values(by='TIR', ascending=False).head(5)
```

---
# 8) Margen de beneficio

    Margen de beneficio = (Beneficio neto anual / prediccion precio aplicado al año) * 100

```python
df_results['Margen de beneficio'] = ((df_results['Beneficio neto anual']/df_results['Prediccion precio alquiler aplicado al año'])*100).round(3)
```

```python
df_results[['Margen de beneficio']].sort_values(by='Margen de beneficio', ascending=False).head(5)
```
